```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('fruits_dataset.csv')
data
```

```python
data.drop(columns = data.columns[0], inplace=True)
df = data.copy()
df
```

```python
piv = df.pivot(index = ['region', 'fruits'], columns= ['sizes'], values= 'units')
piv
```

```python
piv_1 = df.pivot_table(index= ['region', 'fruits'], columns = ['tastes'], values= 'units', aggfunc= sum)
piv_1
```

```python
fruits = ['apple', 'pear', 'pear', 'apple', 'pear']
colors = ['red', 'yellow', 'brown', 'brown', 'green']
tastes = ['sour', 'sour', 'sweet', 'sweet', 'bitter']
sizes = ['small', 'big', 'big', 'big', 'small']
calories = [155, 130, 129, 143, 148]
region = ['north', 'north', 'south', 'north', 'north']
cost = [35, 27, 40, 8, 16]
units = [210, 197, 187, 208, 201]

cols = ['fruits', 'colors', 'tastes', 'sizes', 'calories', 'region', 'cost', 'units']

data = pd.DataFrame(data= [fruits,
                           colors,
                           tastes,
                           sizes,
                           calories,
                           region, 
                           cost,
                           units]).transpose()
data.columns = cols
data
```

```python
df = data.copy()
df
```

```python
piv = df.pivot(index=['colors', 'tastes'], columns= ['region'], values=['units'])
piv
```

```python
piv_1 = df.pivot_table(index=['colors', 'tastes'], columns= ['sizes'], values=['units'], aggfunc= sum)
piv_1
```
