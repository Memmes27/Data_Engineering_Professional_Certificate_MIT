```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('fruits_dataset.csv')
data.drop(columns= data.columns[0], inplace=True)
data
```

```python
pv = data.pivot_table(index=['region', 'fruits'],
              columns= ['tastes', 'sizes'], 
              values= ['units'],
              aggfunc= sum)
pv
```
