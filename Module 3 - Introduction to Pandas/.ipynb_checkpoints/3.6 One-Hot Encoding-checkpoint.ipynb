```python
import pandas as pd
import numpy as np
```

```python
cols = list('ABCD')
np.random.seed(seed=1)
myArr = np.random.ranf(size=(5,4))
df = pd.DataFrame(data=myArr, columns=cols)
df['Fruit'] = ['apple', 'pear', 'orange', 'pear', 'apple']
df
```

- One-hot encoding converts the categorical variables of a data set into a form that is understandable to machine learning algorithims

- pd.get_dummies(df['column']) function converts categorical variables into dummy variables

```python
one_hot = pd.get_dummies(df['Fruit'])
df = df.join(one_hot)
df
```

```python
df.drop('Fruit', axis=1, inplace=True)
df
```
